## Installation for Anaconda Python and Packages

https://www.anaconda.com/products/distribution

For window's installation, click on the download button on the right.

For MACOS/Linux or Win-32 bit OS platform, please click on the icons below "Get Additional Installers"

### Window's Setup.

For MACOS/Linux please refer to https://problemsolvingwithpython.com/01-Orientation/01.04-Installing-Anaconda-on-MacOS/

Open up Anaconda and check if Jupyter Notebook is installed successfully.if you see Jupyter Notebook has a button launch, it means its installed successfully. Otherwise click on the install button

Click on the Window's button, in the search bar, type "Anaconda" and click on Anaconda Prompt (anaconda3)

Follow the remaining instructions in the Anaconda prompt to install the following packages
